# Data Cleaning

```python
# Convert column names to lowercase
df.columns = df.columns.str.lower()
df.columns
```

```python
# Drop unnecessary columns
df.drop(['id', 'birth'], axis= 1, inplace= True)
```

```python
# Check duplicates after dropping columns
df.duplicated().sum()
```

```python
# Convert object columns to numerical
df[['income', 'home_val', 'bluebook', 'oldclaim', 'clm_amt']]
```

```python
# First method to convert from object to float
df['income'].str.replace('$', '').str.replace(',', '').astype('float')
```

```python
def convert_to_numeric(x):

    if type(x) == float:
        return x
    
    else:
        return float(x.strip('$').replace(',', ''))
```

```python
df['income'].apply(convert_to_numeric)
```

```python
# Update the 5 columns

convert_cols = ['income', 'home_val', 'bluebook', 'oldclaim', 'clm_amt']

for col in convert_cols:

    df[col] = df[col].apply(convert_to_numeric)
```

```python
df[['income', 'home_val', 'bluebook', 'oldclaim', 'clm_amt']]
```
